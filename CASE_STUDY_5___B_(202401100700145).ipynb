```python
import pandas as pd
import matplotlib.pyplot as plt

# âœ… Step 2: Upload Excel file manually (if not using Google Drive)
from google.colab import files
uploaded = files.upload()  # Select your Marks.xlsx file here

# âœ… Step 3: Read Excel file
file_path = list(uploaded.keys())[0]  # Gets the uploaded filename
MAX_MARKS = 15
# Explicitly specify the engine for read_excel
df = pd.read_excel(file_path, engine='openpyxl') # Added engine='openpyxl' to handle .xlsx files

# âœ… Step 4: Calculate percentage
df['Percentage'] = (df['Marks'] / MAX_MARKS) * 100

# âœ… Step 5: Categorize students
greater_75 = df[df['Percentage'] > 75]
between_60_75 = df[(df['Percentage'] >= 60) & (df['Percentage'] <= 75)]
less_60 = df[df['Percentage'] < 60]

# âœ… Step 6: Print categorized data
print("ðŸ”· Students with >75%:\n", greater_75)
print("\nðŸ”¶ Students with 60%-75%:\n", between_60_75)
print("\nðŸ”» Students with <60%:\n", less_60)

# âœ… Step 7: Plot Histogram
plt.figure(figsize=(8, 5))
plt.hist(df['Percentage'], bins=range(0, 101, 5), edgecolor='black', color='blue')
plt.xlabel('Percentage')
plt.ylabel('Number of Students')
plt.title('Histogram Plot')
plt.grid(True)
plt.show()

# âœ… Step 8: Plot Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(df['Name'], df['Percentage'], color='green', edgecolor='black')
plt.xlabel('Students')
plt.ylabel('Percentage')
plt.title('Scatter Plot')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
